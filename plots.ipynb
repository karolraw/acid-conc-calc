```python
# The calculations below regarding selected acids were based on their percentage concentrations and the corresponding densities in 20°C.
# The data was taken from this wikipedia page: https://pl.wikibooks.org/wiki/Aneks/G%C4%99sto%C5%9B%C4%87_i_sk%C5%82ad_procentowy_roztwor%C3%B3w).

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

url = r'https://pl.wikibooks.org/wiki/Aneks/G%C4%99sto%C5%9B%C4%87_i_sk%C5%82ad_procentowy_roztwor%C3%B3w'

# Extract and transform data
sulphuric_acid = pd.read_html(url,thousands=None,match='kwasu siarkowego')[0] # DataFrame
X = sulphuric_acid.iloc[:,0].astype(float).to_numpy().reshape(-1,1)
y = sulphuric_acid.iloc[:,1].str.replace(',','.').astype(float).to_numpy()

# Transform the X datapoints to polynomials
poly_features = PolynomialFeatures(degree=6,include_bias=False)
X_poly = poly_features.fit_transform(X)

# Train the model
reg = LinearRegression()
reg.fit(X_poly,y)

y_pred = reg.predict(X_poly)

# Plot the data and prediction
plt.scatter(X,y,label='data points')
plt.plot(X,y_pred,color='g',linestyle='--',label='prediciton')
plt.title('H2SO4')
plt.xlabel('concentration in %')
plt.ylabel('density (20°C) in g/cm3')
plt.legend()

print(f'Fitting score: {reg.score(X_poly,y)}')

# Generate the equation
feature_names = poly_features.get_feature_names_out().tolist()
coefficients = reg.coef_.tolist()
intercept = reg.intercept_

terms = [f"{coeff}*{name}" for coeff, name in list(reversed(list(zip(coefficients, feature_names))))]
equation = " + ".join(terms) + f" + {intercept}"

print('Equation of the model:')
print(equation)
```
